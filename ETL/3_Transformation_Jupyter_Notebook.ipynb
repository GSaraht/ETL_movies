```python
#import libraries
import pandas as pd
import numpy as np
import re
```

```python
#Upload all dfs
#movies from Bechdel websites
dfm = pd.read_csv('1_Bechdel_Test.csv', sep=',', encoding='utf-8', low_memory=False)
dfm.head()
```

```python
#genre from Boxoffice
dfg = pd.read_csv('1_Boxoffice_original_scraped_data.csv', sep=',', encoding='utf-8', low_memory=False)
dfg.head()
```

```python
#Ratings from Cineman
dfr04 = pd.read_csv('2_Cineman_original_scraped_data_part1.csv', sep=',', encoding='utf-8', low_memory=False) #0-4k lines
dfr46 = pd.read_csv('2_Cineman_original_scraped_data_part2.csv', sep=',', encoding='utf-8', low_memory=False) #4-6k lines
dfr68 = pd.read_csv('2_Cineman_original_scraped_data_part3.csv', sep=',', encoding='utf-8', low_memory=False) #6-8k lines
dfr89 = pd.read_csv('2_Cineman_original_scraped_data_part4.csv', sep=',', encoding='utf-8', low_memory=False) #8-9k lines
dfr04.head()
```

```python
#merge/ concat the ratings dfs into one ratings Csv
ratings = [dfr04, dfr46, dfr68, dfr89]
dfr = pd.concat(ratings)
dfr.head()
```

```python
len(dfr) #7815 rows
```

```python
#Merge all dfs with movie_title
df = dfm.merge(dfg, on='title').merge(dfr, on='title')
#df.head()
len(df) #10019 rows
```

```python
#Looking for some strange expressions which are/were present
print(dfm.iloc[[766]]) #TrÊs 
print(dfm.iloc[[5275]]) #&amp
```

```python
#Adjust some content in column title
df['title'] = df['title'].replace('&amp;','&', regex=True)
df['title'] = df['title'].replace("TrÊs ", "Três", regex=True)
#print(dfm.iloc[[766]]) #TrÊs 
#print(dfm.iloc[[5275]]) #&amp
```

```python
#Replaceing 'movie not found' with 'nan'
df['genre'] = df['genre'].replace("movie not found", str('NaN')) 
df.head()
```

```python
#Replacing empty space between the genre tags
df['genre'] = df['genre'].str.replace(' ',', ')
df.head()
```

```python
#Replaceing 'no score' with 'nan'
df['review_by_cineman'] = df['review_by_cineman'].replace("no score", str('NaN')) 
df['cineman_movie_charts'] = df['cineman_movie_charts'].replace("no score", str('NaN')) 
df.head()
```

```python
#Replaceing 'couldn't find movie' with 'nan'
df['review_by_cineman'] = df['review_by_cineman'].replace("couldn't find movie", str('NaN')) 
df['cineman_movie_charts'] = df['cineman_movie_charts'].replace("couldn't find movie", str('NaN')) 
df.head()
```
